## 특수문자 제거

정규 표현식을 이용하여 한글 영어 숫자 _ - + 띄어어쓰기기 남기고 제거

```python
df['product_title_regex'] = df['product_title'].str.replace(r'[^ㄱ-ㅎㅏ-ㅣ가-힣a-zA-Z0-9-_+ ]', '', regex=True)
```

# 한국어 띄어쓰기 패키지(pykospacing)

```python
from pykospacing import Spacing
```

```python
spacing = Spacing()
```

```python
def spacing_title(x):
    kospacing_sentence = spacing(x)
    return kospacing_sentence
```

```python
df['product_title_kospacing'] = df.product_title_regex.map(lambda x: spacing_title(x))
```

```python
df[df.product_title_regex != df.product_title_kospacing][['product_title_regex', 'product_title_kospacing']].sample(20)
```

## 불용어 제거

```python
stop_words = ['팝니다', '팔아요', '팜', '판매', '삽니다', '사요', '삼']
```

```python
def remove_stop_words(x):
    result = [word for word in x.split() if not word in stop_words]
    return ' '.join(result)
```

```python
df['product_title_remove_stop_words'] = df.product_title_kospacing.map(lambda x: remove_stop_words(x))
```

```python
df
```

# Tokenization(Pynori)

```python
from pynori.korean_analyzer import KoreanAnalyzer
```

```python
nori = KoreanAnalyzer(
           decompound_mode='DISCARD', # DISCARD or MIXED or NONE
           infl_decompound_mode='DISCARD', # DISCARD or MIXED or NONE
           discard_punctuation=True,
           output_unknown_unigrams=False,
           pos_filter=False,
           synonym_filter=False, mode_synonym='NORM', # NORM or EXTENSION
       ) 
```

```python
def analyzer(x):
    result = nori.do_analysis(x)
    return result['termAtt'], result['posTagAtt']
```

```python
test = df.copy()
df = test.sample(10)
```

```python
df['product_title_nori_result'], df['product_title_nori_tags']  = zip(*df.product_title_remove_stop_words.map(lambda x: analyzer(x)))
```
