```python
from sklearn.datasets import load_digits

digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31]

d =[digits.images[samples[i]] for i in range(len(samples))]

plt.figure(figsize=(8, 2))

for i in range(len(d)):
    plt.subplot(1, 8, i + 1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'image {i + 1}')
plt.suptitle('숫자 0과 1 이미지')
plt.tight_layout()
plt.show()
```

#### 2차원 이미지 데이터를 1차원 벡터로 변환

```python
vectors = [d[i].reshape(64, 1) for i in range(len(d))] # reshape를 이용하여 벡터화

plt.figure(figsize=(8, 3))
for i in range(len(vectors)):
    plt.subplot(1, 8, i + 1)
    plt.imshow(vectors[i], aspect=0.4,interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'벡터 {i + 1}')
plt.suptitle('벡터화된 이미지', y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

#### 행렬(matrix)
- 복수의 차원을 가지는 데이터 레코드가 다시 여러개 있는 경우의 데이터
- 열 벡터(column vector): 열의 수가 1인 행렬
- 하나의 데이터 레코드를 단독으로 벡터로 나타낼 때는 하나의 열(column)로 나타내고 복수의 데이터 레코드 집합을 행렬로 나타낼때는 하나의 데이터 레코드가 하나의 행(row)으로 표기

```python
A = np.array([[11, 12, 13], [21, 22, 23]])
A
```

#### 전치 연산
- 행렬에서 가장 기본이 되는 연산
- 행과 열을 바꾸는 연산
- 벡터나 행렬에 T라는 위첨자를 붙여서 표기
- 전치 연산으로 만든 행렬을 원래 행렬에 대한 전치행렬이라고함
- 벡터(열 벡터)에 대해 전치 연산을 적용하면 행의 수가 1인 행렬이므로 행 벡터라고 함

```python
A = np.array([[11, 12, 13], [21, 22, 23]])
A
```

```python
# numpy에서 전치 연산 
A.T
```

```python
# 1차원은 전치 연산이 정의되지 않음
x1 = np.array([5.1, 3.5, 1.4, 0.2])
x1, x1.T
```

### 특수한 벡터와 행렬

#### 영벡터(zeros-vector)
- 모든 원소가 0인 N차원 벡터

```python
# numpy에서 영벡터는 zeros()로 만듬
zeros_vector = np.zeros((3, 1)) # 3열 1행의 열 벡터
zeros_vector
```

#### 일벡터(ones-vector)
- 모든 원소가 1인 N차원 벡터

```python
# numpy에서 영벡터는 ones()로 만듬
ones_vector = np.ones((3, 1)) # 3열 1행의 열 벡터
ones_vector
```

#### 정방 행렬(square matrix)
- 행의 개수와 열의 개수가 같은 행렬

```python
square_matrix = np.array([[1, 2], [3, 4]])
square_matrix.shape
```

#### 대각 행렬(diagonal matrix)
- 행렬에서 행과 열이 같은 위치를 대각 이라고함
- 대각 위치에 있지않은 것들은 비대각 이라고함
- 모든 비대각 요소가 0인 행렬을 대각 행렬이라고함
- 대각 행렬이 되려면 비대각 성분이 0이기만 하면 되고 대각성분은 0이든 아니든 상관없음
- 정방행렬일 필요없음

```python
# numpy에서 정방 행렬은 diag()로 만듬
np.diag([1, 2, 3])
```

#### 항등행렬(identity matrix)
- 대각행렬 중에서도 모든 대각성분의 값이 1인 대각 행렬

```python
# numpy에서 항등행렬은 identity() 혹은 eye()로 만듬
np.identity(3)
```

```python
np.eye(4)
```

#### 대칭행렬(symmetric matrix)
- 전치연산을 통해 얻은 전치행렬과 원래 행렬이 같으며 대칭행렬

```python
A = np.array([[1, 2], [2, 1]])
A, A.T
```
